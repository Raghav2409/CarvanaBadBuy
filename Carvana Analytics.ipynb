```python
#Make Vs 'IsBadBuy'
grouped_data = data.groupby('Make')['IsBadBuy'].mean() * 100

#bar plot
fig, ax = plt.subplots(figsize=(50,10))
ax.bar(grouped_data.index, grouped_data)

#Labels
ax.set_xlabel('Make')
ax.set_ylabel('Mean IsBadBuy Percentage')
ax.set_title('Mean IsBadBuy Percentage by Make')

plt.show()
```

```python
#'MMRAcquisitionAuctionAveragePrice', 'MMRAcquisitionAuctionCleanPrice', 'MMRAcquisitionRetailAveragePrice', 'MMRAcquisitonRetailCleanPrice' Vs IsBadBuy

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))

# regular scale
ax = data.boxplot(column='MMRAcquisitionAuctionAveragePrice', by='IsBadBuy', ax=axes[0][0])
ax.set_ylabel('MMRAcquisitionAuctionAveragePrice')
ax.set_xlabel('IsBadBuy')


ax = data.boxplot(column='MMRAcquisitionAuctionCleanPrice', by='IsBadBuy', ax=axes[0][1])
ax.set_ylabel('MMRAcquisitionAuctionCleanPrice')
ax.set_xlabel('IsBadBuy')

ax = data.boxplot(column='MMRAcquisitionRetailAveragePrice', by='IsBadBuy', ax=axes[1][0])
ax.set_ylabel('MMRAcquisitionRetailAveragePrice')
ax.set_xlabel('IsBadBuy')

ax = data.boxplot(column='MMRAcquisitonRetailCleanPrice', by='IsBadBuy', ax=axes[1][1])
ax.set_ylabel('MMRAcquisitonRetailCleanPrice')
ax.set_xlabel('IsBadBuy')


# suppress the title
#axes[0].get_figure().suptitle('')
plt.tight_layout()

plt.show()
```

```python
#'MMRCurrentAuctionAveragePrice', 'MMRCurrentAuctionCleanPrice', 'MMRCurrentRetailAveragePrice', 'MMRCurrentRetailCleanPrice' Vs IsBadBuy
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))

# regular scale
ax = data.boxplot(column='MMRCurrentAuctionAveragePrice', by='IsBadBuy', ax=axes[0][0])
ax.set_ylabel('MMRCurrentAuctionAveragePrice')
ax.set_xlabel('IsBadBuy')

ax = data.boxplot(column='MMRCurrentAuctionCleanPrice', by='IsBadBuy', ax=axes[0][1])
ax.set_ylabel('MMRCurrentAuctionCleanPrice')
ax.set_xlabel('IsBadBuy')

ax = data.boxplot(column='MMRCurrentRetailAveragePrice', by='IsBadBuy', ax=axes[1][0])
ax.set_ylabel('MMRCurrentRetailAveragePrice')
ax.set_xlabel('IsBadBuy')

ax = data.boxplot(column='MMRCurrentRetailCleanPrice', by='IsBadBuy', ax=axes[1][1])
ax.set_ylabel('MMRCurrentRetailCleanPrice')
ax.set_xlabel('IsBadBuy')


# suppress the title
#axes[0].get_figure().suptitle('')
plt.tight_layout()

plt.show()
```

```python
#Profit/Loss Vs IsBadBuy
profitloss = data['MMRCurrentRetailCleanPrice'] - data['MMRAcquisitonRetailCleanPrice']
data['profitloss'] = profitloss

grouped_data3 = data.groupby('IsBadBuy')['profitloss'].mean() * 100

# Create a bar plot
fig, ax = plt.subplots(figsize=(30,6))
ax.plot(grouped_data3.index, grouped_data3)

# Label the plot
ax.set_xlabel('VNST')
ax.set_ylabel('Mean IsBadBuy Percentage')
ax.set_title('Mean IsBadBuy Percentage by Transmission')

# Show the plot
plt.show()
```
